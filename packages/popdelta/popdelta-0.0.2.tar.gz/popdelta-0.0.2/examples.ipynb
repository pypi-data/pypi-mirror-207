# Non Weighted 1 Population - Pop Delta 

```python
from popdelta.pop_delta import PopDelta

# NW1
NW1_target = ""
NW1_num_bins = 3
NW1_string_attributes = []

popDeltaNW1 = PopDelta(target=NW1_target, num_bins=NW1_num_bins, string_attributes=NW1_string_attributes)

k=0
for e in popDeltaNW1.process_batch([A]):
    k +=1
    print("iteration", k)

display(e)
```

# Weighted 1 Population - Pop Delta 

```python
from popdelta.pop_delta import PopDelta

# W1
W1_target = "age"
W1_num_bins = 3
W1_string_attributes = []

popDeltaW1 = PopDelta(target=W1_target, num_bins=W1_num_bins, string_attributes=W1_string_attributes)

k=0
for e in popDeltaW1.process_batch([A]):
    k +=1
    print("iteration", k)

display(e)
```

# Non Weighted 2 Population - Pop Delta 

```python
from popdelta.pop_delta import PopDelta

# NW2
NW2_target = ""
NW2_num_bins = 3
NW2_string_attributes = []

popDeltaNW2 = PopDelta(target=NW2_target, num_bins=NW2_num_bins, string_attributes=NW2_string_attributes)

k=0
for e in popDeltaNW2.process_batch([A, B]):
    k +=1
    print("iteration", k)

display(e)
```
