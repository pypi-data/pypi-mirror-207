```python
from terminal_tracker import Tags

tg = Tags("history_files/zsh_history.txt", timeframe=False, shell="zsh")

# Search for "project", get the dataframe
nlp_df = tg.search_df("NLP")

nlp_df.head(10)
```

```python
# Search for "project", get the commands
project_command_list = tg.search("NLP")
for command in set(project_command_list):
    print(command)
```
